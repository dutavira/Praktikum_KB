```python
import pandas as pd
```

Pada cell pertama menggunakan import library pandas supaya dapat mengelola dataframe dari file csv

```python
covid = pd.read_csv("data_covid.csv")
```

Cell kedua terdapat variabel yang isinya berfungsi untuk membaca file csv

```python
jumlahRecord = covid.shape[0]
Atribut = []
for row in covid:
    Atribut.append(row)
tipeObjek = list(covid.select_dtypes(include=['object']))
tipeInt = list(covid.select_dtypes(include=['int64']))
tipeNull = list(covid.columns[covid.isna().sum() > 0])
JumAtributObj= len(tipeObjek)
JumAtributInt = len(tipeInt)
JumAtributNull = len(tipeNull)

```

Pada cell ketiga
baris pertama = berfungsi untuk menampilkan jumlah atribut yang terdapat pada data di file csv
baris kedua = terdapat variabel Atribut yang berisi list kosong
baris ketiga - keempat = berisi perulangan untuk memasukan kolom yang ada pada data csv ke dalam vaiabel Atribut
baris kelima = terdapat variabel list dengan isinya berupa atribut yang bertipe 'object'
baris keenam = terdapat variabel list dengan isinya berupa atribut yang bertipe 'int64'
baris ketujuh = terdapat variabel list dengan isinya berupa atribut yang bertipe 'null'
baris kedelapan - ke 10 = terdapat varibel yang isinya untuk menghitung panjang dari hasil variabel tipeObjek, tipeINT, dan tipeNull

```python
rekap = {'Jumlah Record' : jumlahRecord,
        'Nama Attribute': Atribut,
        'Jumlah Attribute(Objek)': JumAtributObj,
        'Jumlah Attribute(Int)': JumAtributInt,
        'Jumlah Attribute(NULL)':JumAtributNull}
print('Jumlah Record', "\t\t: ", rekap['Jumlah Record'], "\n")
print('Nama Atribut', "\t\t: ", rekap['Nama Attribute'], "\n")
print('Jumlah Attribute(Objek)', ": ", rekap['Jumlah Attribute(Objek)'], "\n")
print('Jumlah Attribute(Int)', "\t: ", rekap['Jumlah Attribute(Int)'], "\n")
print('Jumlah Attribute(NULL)', "\t: ", rekap['Jumlah Attribute(NULL)'], "\n")
```
