import library terlebih dahulu

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

memanggil database, disini saya menggunakan dua jenis database

```python
df_timeseries = pd.read_csv("Dataset/data_covid.csv")
df_categorical = pd.read_csv('Dataset/california_house_price.csv')
```

membuat grafik batang

```python
# plt.rcParams.update({'font.size': 5})
plt.barh(df_timeseries['Province'], df_timeseries['Daily_Case'])
plt.title("Kasus Harian Covid-19 di Indonesia")
plt.ylabel("Provinsi")
plt.xlabel("Jumlah Kasus")
plt.show 
```

grafik batang diatas dapat menunjukkan jumlah kasus harian terbanyak terjadi di DKI JAKARTA sedangkan yang paling sedikit terjadi di SULAWESI BARAT

Membuat grafik plot

```python
tanggal = [item for item in range(0, len(df_timeseries['Date']))]
plt.plot(tanggal, df_timeseries['Daily_Death'], color='blue', linewidth=0.5)
plt.legend(['Kasus Meninggal'])
plt.title('Data Covid-19 di Indonesia')
plt.xlabel('Hari ke-')
plt.ylabel('Jumlah')
```

pada grafik plot diatas menggambarkan jumalah kasus kematian perhari akibat Covid di Indonesia, dimulai pada hari ke 0 dan mengalami kenaikan sampai sebelum hari ke 2500

membuat grafik pie

```python
labels = df_categorical['ocean_proximity'].value_counts().rename_axis("Area").reset_index(name='Jumlah')
plt.figure(figsize=(5, 5), dpi=100)
plt.pie(df_categorical['ocean_proximity'].value_counts(), labels=labels['Area'], autopct='%.1f%%', startangle=90)
plt.title("Area Rumah Dijual di California")
plt.show
```

grafik pie diatas menggambarkan presentase jumlah rumah dijual pada tiap area di California

membuat grafik boxplot

```python
sns.boxplot(x=df_categorical['total_rooms'], y=df_categorical['ocean_proximity'])
plt.show()
```

Grafik boxplot di atas menggambarkan persebaran data dari total room di setiap area(ocean_proximity) California
